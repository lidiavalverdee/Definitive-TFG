```python

# Definir los valores para la nueva fila
valores= ['Type of accident', 'Gender', 'Nationality', 'Professional situation', 'Main occupation', 'Seniority in the job (months)', 'Seniority in the job (days)', 'Type of contract', 'Social Security Scheme', 'Economic activity of the centre', 'Centre staff (victime)', 'Province (victime)', 'Contracts or subcontracts', 'Temporary employment', 'Personal assumption', 'Own service', 'External prevention service', 'Designated workers', 'Pooled service', 'Preventive Org: None', 'Accident location', 'Traffic accident', 'The centre belongs to the company where the worker is affiliated to the Social Security', 'The centre is the same as the centre where the worker is affiliated', 'Contracts or subcontracts the company', 'Temporary employment user', 'Another employer-employee relationship', 'Province (accident place)', 'Centre staff (place)', 'Economic activity of the centre', 'Date of accident', 'Date of medical leave', 'Weekday', 'Time of day', 'Working time', 'Regular job', 'Risk assessment', 'Type of place', 'Type of work', 'Specific physical activity', 'Material agent associated with the physical activity', 'Deviation that triggered the accident', 'Material agent associated with the deviation', 'Form or contact that produced the injury', 'Material agent associated with the form or contact', 'Affected more than one worker', 'Description of injury', 'Degree of injury', 'Part of body injured', 'Type of assistance', 'Hospitalisation', 'Daily amount of benefit', 'Degree of injury (at the time of discharge)', 'Date of medical discharge', 'Medical diagnosis code (at the time of discharge)', 'Duration of sick absence', 'Age of the injured worker on the day of the accident', 'Province of the Labour Authority receiving the occupational accident report']  # Reemplaza estos valores con los que desees añadir
datos.columns = valores

datos
```

```python

# Filtrar las filas donde el valor en la columna 'province' sea igual a 8 (Barcelona)
datos_filtrados = datos[datos.iloc[:, 11] == 8]

datos_filtrados= datos_filtrados.reset_index(drop=True)

# Mostrar las filas filtradas
datos_filtrados.iloc[:, :12]

# Guardar el DataFrame resultante en un nuevo archivo CSV
datos_filtrados.to_csv('DadesMinisteriBarcelona.csv', index=False)

datos_filtrados


```

```python
# Contar el número de hombres (valor 1) y mujeres (valor 2) en la columna 1
conteo_genero = datos_filtrados.iloc[:, 0].value_counts()

print(conteo_genero)
```

```python
# Crear un gráfico de barras
conteo_genero.plot(kind='bar')

# Etiquetas y título del gráfico
plt.xlabel('Género')
plt.ylabel('Número de personas')
plt.title('Número de hombres y mujeres')

# Mostrar el gráfico
plt.show()
```

```python
# Contar el número de accidentes para cada valor único en cada columna (buscamos la columna por el nombre de la variable '')
variable= 'Specific physical activity' #canviar per a cada variable
conteo_accidentes = datos_filtrados[variable].value_counts()

print(conteo_accidentes)
print(type(conteo_accidentes))

conteo_accidentes=dict(conteo_accidentes)

nombres_barras = list(conteo_accidentes.keys())
conteos = list(conteo_accidentes.values())

# Diccionario de mapeo CAMBIA PARA CADA VARIABLE
mapeo={}
#mapeo = {1: 'Accident', 2: 'Relapse'}
#mapeo = {1: 'Men', 2: 'Women'}
#mapeo = {1 : 'Private sector employees', 2 : 'Public sector employees', 3 : 'Self-employed with employees', 4 : 'Self-employed without employees'}
#mapeo = {1: 'Usual Centre', 2: 'On the move', 3: 'In itinere', 4: 'Another location'}
#mapeo = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7:'Sunday'}
#mapeo = {0: 'No', 1: 'Yes'}
#mapeo = {1:'Low', 2: 'Serious', 3: 'Very serious', 4: 'Lethal' }

# Aplica el mapeo a la lista usando una comprensión de lista
lista_actualizada = [mapeo.get(valor, valor) for valor in nombres_barras]

print(lista_actualizada)  # Esto imprimirá: ['Accidente', 'Recaída']

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(range(len(conteo_accidentes)), conteos)
#conteo_accidentes.plot(kind='bar')
#plt.title('Número de Accidentes por Valor de la Columna 2')

plt.xticks(range(len(conteo_accidentes)), lista_actualizada, rotation=45)

plt.xlabel(variable)
plt.ylabel('Number of accidents')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mayor claridad
plt.grid(axis='y')  # Agregar rejilla en el eje y
plt.show()

```
